## import sqllite

```python
import sqlite3 
import pandas as pd
```

## Koneksi

```python
def Konek():
    return sqlite3.connect("data_penjualan_ori.db")
```

## Soal No 1

#### Tampilkan data order yang mempunyai order amount di bawah 1000

```python
def Tampil_data(con):
    connection = con.cursor()
    connection.execute("SELECT name from sqlite_master where type= 'table'")
    print(connection.fetchall())
```

```python
koneksi = Konek()
Tampil_data(koneksi)
```

## Tampil FROM DATAFRAME

```python
connection = Konek()

# Buat Query SQL
result  = connection.cursor()
command = "SELECT * FROM orders WHERE ORD_AMOUNT < '1000.00';"

# Eksekusi Query SQL
result.execute(command)


results      = result.fetchall();connection.close();
cols         = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)

results_data
```

## Soal no 4

#### Tampilkan informasi customer yang mempunyai payment amount diatas 8000 dan urutkan berdasarkan nama customer dari A sampai Z

## Soal no 5

#### Tampilkan data customer yang tidak tinggal di USA dan memiliki grade lebih dari 1
