```python
import sqlite3
import pandas as pd
```

### Koneksi Ke Database

```python
def koneksi():
    return sqlite3.connect("databaseku.db")
def koneksi_penjualan():
    connection = sqlite3.connect("data_penjualan_ori.db")
    return connection
def koneksi_siswa():
    return sqlite3.connect("data_siswa.db")
```

```python
def Tampil_Table_siswa(con):
    cn = con.cursor()
    cn.execute("SELECT * FROM siswa")
    print(cn.fetchall())
```

```python
def Tampil_table_mahasiswa(con):
    cn = con.cursor()
    cn.execute("SELECT * FROM mahasiswa")
    print(cn.fetchall())
```

### Koneksi Ke Databaseku

```python
#menampilkan table siswa
conn = koneksi()
Tampil_Table_siswa(conn)
```

```python
#menampilkan table mahasiswa
conn = koneksi()
Tampil_table_mahasiswa(conn)
```

### Create Table Siswa Di dalam Database databaseku

```python
conn = koneksi()
result = conn.cursor()
sql = """
            CREATE TABLE siswa (
               id_siswa INTEGER PRIMARY KEY,
               nama TEXT NOT NULL,
               umur INTEGER NOT NULL
            );
        """
result.execute(sql)
conn.close()
```

```python
def lihat_daftar_tabel(con):
    cn = con.cursor()
    cn.execute('SELECT name from sqlite_master where type= "table"')
    print(cn.fetchall())
```

```python
conn = koneksi()
lihat_daftar_tabel(conn)
```

```python
conn = koneksi_penjualan()
lihat_daftar_tabel(conn)
```
